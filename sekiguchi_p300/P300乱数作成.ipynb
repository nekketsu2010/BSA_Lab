```python
import numpy as np
import random
import itertools

a = []
for num in range(30):
    aa = []
    for n in range(12):
        for i in range(1):
            aa.append(n)
    a.append(aa)
    
len(a), a
```

```python
ok = []

ok.append(random.sample(a[0], len(a[0])))

for i in range(1, len(a)):
    temp = random.sample(a[i], len(a[i]))
    while ok[i-1][-1] == temp[0]:
        temp = random.sample(a[i], len(a[i]))
    ok.append(temp)

ok
```

```python
result = list(itertools.chain.from_iterable(ok))

a = result[1:]
b = result[:-1]

n = 0
for i in range(len(a)):
    if a[i] == b[i]:
        n += 1
print(n)

len(result), result
```

```python
def generate_random():
    a = []
    for num in range(30):
        aa = []
        for n in range(12):
            for i in range(1):
                aa.append(n)
        a.append(aa)

    ok = []

    ok.append(random.sample(a[0], len(a[0])))

    for i in range(1, len(a)):
        temp = random.sample(a[i], len(a[i]))
        while ok[i-1][-1] == temp[0]:
            temp = random.sample(a[i], len(a[i]))
        ok.append(temp)

    result = list(itertools.chain.from_iterable(ok))

    a = result[1:]
    b = result[:-1]

    n = 0
    for i in range(len(a)):
        if a[i] == b[i]:
            n += 1
    print(n)
    
    return result
```

```python
for i in range(100):
    with open("exp_img_random_show" + str(i), 'wb') as f:
        pickle.dump(generate_random(), f)
```
