```python
csv = pd.read_csv("201103/alpha_wave_03.11.20_23.18.11.md.csv", skiprows=1)
csv
```

```python
O1 = csv['EEG.O2'].values
O1
```

```python
plt.figure(figsize=(16, 8))
# tt = np.linspace(0, 128, AF3.shape[0])
plt.plot(O1)
plt.xlim(0, 5120)
plt.show()
```

```python
O1_FFT = np.abs(np.square(np.fft.fft(O1)))
plt.figure(figsize=(16, 8))
tt = np.linspace(0, 128, O1_FFT.shape[0])
plt.plot(tt, O1_FFT)
plt.xlim(0, 64)
plt.ylim(0, 0.3*10**9)
plt.xlabel("Frequency[Hz]")
plt.ylabel("Amplitude")
plt.show()
```

```python
from scipy import signal
Fs = 128
# firフィルタを適用する
def firFilter(x):
    # fir filter
    b = signal.firwin(Fs * 2 - 1, [1.5/(Fs/2), 15.0/(Fs/2)], pass_zero=False)
    x = signal.filtfilt(b, 1, x)
    return x
```

```python
O1 = firFilter(O1)
O1
```

```python
plt.figure(figsize=(16, 8))
# tt = np.linspace(0, 128, AF3.shape[0])
plt.plot(O1)
plt.xlim(12000, 13500)
# plt.xticks(np.arange(0, O1.shape[0]+1, 1000))
plt.show()
```

```python
O1_FFT = np.abs(np.square(np.fft.fft(O1[19200:23040])))
plt.figure(figsize=(16, 8))
tt = np.linspace(0, 128, O1_FFT.shape[0])
plt.plot(tt, O1_FFT)
plt.xlim(0, 15)
# plt.ylim(0, 0.3*10**7)
plt.xlabel("Frequency[Hz]")
plt.ylabel("Amplitude")
plt.show()
```

```python
O1_FFT = np.abs(np.square(np.fft.fft(O1[12000:15840])))
plt.figure(figsize=(16, 8))
tt = np.linspace(0, 128, O1_FFT.shape[0])
plt.plot(tt, O1_FFT)
plt.xlim(0, 15)
# plt.ylim(0, 2.0*10**6)
plt.xlabel("Frequency[Hz]")
plt.ylabel("Amplitude")
plt.show()
```
