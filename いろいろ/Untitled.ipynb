```python
import numpy as np
import pandas as pd

import seaborn as sns
sns.set(font='Yu Gothic')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
csv = pd.read_csv("emotiv_epoc_plus.csv")
csv
```

```python
AF3 = csv['AF3'].values
AF3
```

```python
plt.figure(figsize=(16, 8))
# tt = np.linspace(0, 128, AF3.shape[0])
plt.plot(AF3)
plt.xlim(0, 256)
plt.show()
```

```python
AF3_FFT = np.abs(np.square(np.fft.fft(AF3)))
plt.figure(figsize=(16, 8))
tt = np.linspace(0, 128, AF3_FFT.shape[0])
plt.plot(tt, AF3_FFT)
plt.xlim(0, 64)
plt.ylim(0, 0.3*10**7)
plt.xlabel("Frequency[Hz]")
plt.ylabel("Amplitude")
plt.show()
```
