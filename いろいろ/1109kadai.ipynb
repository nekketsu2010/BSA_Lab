```python
import numpy as np
import pandas as pd

import seaborn as sns
sns.set(font='Yu Gothic')
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = np.loadtxt("data1109.csv", delimiter=',')
data
```

```python
plt.figure(figsize=(16, 8))
plt.plot(data[:, 0], data[:, 1])
```

```python
x_sum = 0
xt = 0
for i in range(data.shape[0]):
    x_sum += np.square(data[i, 0])
    xt += data[i, 0] * data[i, 1]

w = xt / x_sum
w = w/2

x_sum, xt, w
```

```python
wx = w * (data[:, 0] - np.mean(data[:, 0]))
wx
```

```python
plt.figure(figsize=(16, 8))
plt.plot(data[:, 0], data[:, 1])
plt.plot(data[:, 0], wx + np.mean(data[:, 1]))
```
